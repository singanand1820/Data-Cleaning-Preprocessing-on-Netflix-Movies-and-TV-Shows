```python
import pandas as pd
```

```python
# Load the Netflix Movies and TV Shows dataset
df = pd.read_csv('netflix_titles.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
# Identify missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Handle missing values
# Example: Fill missing 'rating' with 'Not Rated'
df['rating'] = df['rating'].fillna('Not Rated')
```

```python
# Remove duplicate rows
print("\nNumber of duplicate rows:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
# Standardize text columns, Convert 'type' column to lowercase
df['type'] = df['type'].str.lower()

# Standardize 'country' column: lowercase and strip spaces
df['country'] = df['country'].str.lower().str.strip()
```

```python
# Convert date formats to consistent datetime type
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
# Rename columns to lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# Fix data types if necessary, convert 'release_year' to integer
df['release_year'] = df['release_year'].astype(int)
```

```python
# Final checks
print("\nData Types after conversion:")
print(df.dtypes)

print("\nFinal shape after cleaning:", df.shape)
```

```python
# Step 11: Save the cleaned dataset
df.to_csv('Netflix_Titles_cleaned.csv', index=False)

print("\nData cleaning completed and saved as 'Netflix_Titles_cleaned.csv'.")
```
